```python
import os
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
df_start = pd.read_csv("turkish_toxic_language.csv",sep=',')
```

```python
df_start.drop('source',axis=1, inplace=True)
```

```python
df = pd.get_dummies(df_start, columns= ['target'],dtype=int)
```

```python
df = df.rename(columns = {'target_INSULT': 'INSULT', 'target_PROFANITY': 'PROFANITY' , 'target_RACIST':'RACIST','target_SEXIST':'SEXIST'})
```

```python
df
```

```python
from tensorflow.keras.layers import TextVectorization
```

```python
df.columns
```

```python
X = df['text']
y = df[df.columns[1:]].values
```

```python
MAX_FEATURES = 200000 
```

```python
vectorizer = TextVectorization(max_tokens=MAX_FEATURES,
                               output_sequence_length = 1800,
                               output_mode = 'int')
```

```python
vectorizer.adapt(X.values)
```

```python
vectorized_text = vectorizer(X.values)
```

```python
vectorizer('Phil')
```

```python
vectorized_text
```

```python
#MCSHBAP - map, chache, shuffle, batch, prefetch  from_tensor_slices, list_file
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text, y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8) # helps bottlenecks
```

```python
type(dataset)
```

```python
train = dataset.take(int(len(dataset)*.7))
val = dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
test = dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional, Dense, Embedding
```

```python
model = Sequential()
# Create the embedding layer 
model.add(Embedding(MAX_FEATURES+1, 32))
# Bidirectional LSTM Layer
model.add(Bidirectional(LSTM(32, activation='tanh')))
# Feature extractor Fully connected layers
model.add(Dense(128, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
# Final layer 
model.add(Dense(4, activation='sigmoid'))
```

```python
model.compile(loss='BinaryCrossentropy', optimizer='Adam')
```

```python
y[0]
```

```python
history = model.fit(train, epochs=1, validation_data=val)
```
